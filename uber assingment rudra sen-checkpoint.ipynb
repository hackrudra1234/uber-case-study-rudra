```python
#import all things
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
```

```python
#import the uber data and named it as uber
uber = pd.read_csv(r"E:\upgrad\statistics and eda\eda\uber case study\Uber Request Data.csv")
uber.head(10)
```

```python
#getting info about rows and columns
uber.info()
uber.describe()
```

```python
#chekking null values in column
round(100*(uber.isnull().sum()/len(uber.index)),2)
```

```python
#converting to datetime
uber["Request timestamp"] = pd.to_datetime(uber["Request timestamp"],dayfirst = True)
uber["Drop timestamp"] = pd.to_datetime(uber["Drop timestamp"],dayfirst = True)
```

```python
#deriving new columns according to need 
uber['request_date'] = uber["Request timestamp"].dt.date
uber['request_time'] = uber['Request timestamp'].dt.time
uber['drop_time'] = uber["Drop timestamp"].dt.time
uber['drop_date'] = uber["Drop timestamp"].dt.date
```

```python
#converting to date time and extracting day and hours
uber['request_date'] = pd.to_datetime(uber['request_date'],dayfirst = True)
uber["drop_date"] = pd.to_datetime(uber["drop_date"],dayfirst = True)
uber["request_day"] = uber["Request timestamp"].dt.day
uber["request_hour"] = uber["Request timestamp"].dt.hour
uber.head(10)
```

```python
#deriving new columns and mapping week days
uber["request_weeks"] = uber["Request timestamp"].dt.day_name()
```

```python
#droping columns
uber = uber.drop(['Request timestamp','Drop timestamp'],axis = 1)
```

```python
#filling drop date with request date as request date and drop date is same and then cheking data
uber['drop_date'] = uber['drop_date'].fillna(uber['request_date'])
uber.info()
```

```python
#drpping duplicates
uber = uber.drop_duplicates()
```

```python
#making new columns time slot and defining it
uber['time_slot'] = np.where(uber['request_hour'] < 6, "early_morning",np.where(uber['request_hour'] < 12, "day_rush", np.where(uber['request_hour'] < 17,"afternoon",np.where(uber['request_hour'] < 22,"late_evening","late_night"))))
uber.head()
```

```python
#making a new col demand supply and defining it 
uber["demand_supply"] = np.where(uber["Status"] == "Trip Completed","request completed","request denied")
uber
```

```python
#plot for supply and gap
sns.set_style("white")
sns.catplot(x = "demand_supply",kind = "count",data = uber)
plt.title("supply and gap")
plt.ylabel("count of demand and gap")
plt.show()
```

```python
#plot of most pressing problem
pressing_problem = uber.loc[(uber.Status != 'Trip Completed')]
sns.catplot(x = "Status",kind = "count",data = pressing_problem)
plt.title("pressing problem")
plt.show()
```

```python
#plot of demand and supply in frequency of hours
plt.figure(figsize =  (10,8))
sns.countplot(x = "request_hour",hue = "demand_supply",data = uber)
plt.ylabel("count of demand supply")
plt.show()
```

```python
#counting by weeks
sns.catplot(x = "request_weeks",hue = "demand_supply",kind = "count",data = uber)
plt.show()
```

```python
#identifying most gap in timeslot
sns.catplot(x = "time_slot",hue = "demand_supply",kind ="count",data = uber)
plt.show()
```

```python
#identifying most gap pickup point wise
sns.catplot(x = "Pickup point",hue = "demand_supply",kind ="count",data = uber)
plt.show()
```

```python
#identifying most gap in time slot pick point wise
sns.catplot(x = "time_slot",hue = "demand_supply",col = "Pickup point",kind ="count",data = uber)
plt.show()
```

```python
#identifying type of request in identified time slot
status = uber.groupby(["Pickup point","demand_supply","time_slot"])
status["demand_supply"].count()
```
